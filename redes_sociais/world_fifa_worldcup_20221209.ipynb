```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = bd.read_table(dataset_id='world_fifa_worldcup',
table_id='players',
billing_project_id="dados-bd-359914")
```

```python
df_2018 = df[df['list_tournaments'].str.contains('2018')]
df_2014 = df[df['list_tournaments'].str.contains('2014')]
df_2010 = df[df['list_tournaments'].str.contains('2010')]
df_2006 = df[df['list_tournaments'].str.contains('2006')]
df_2002 = df[df['list_tournaments'].str.contains('2002')]
df_1998 = df[df['list_tournaments'].str.contains('1998')]
df_1994 = df[df['list_tournaments'].str.contains('1994')]
df_1990 = df[df['list_tournaments'].str.contains('1990')]
df_1986 = df[df['list_tournaments'].str.contains('1986')]
df_1982 = df[df['list_tournaments'].str.contains('1982')]
```

```python
df_2018["year"]  = pd.DatetimeIndex(df_2018["birth_date"]).year
df_2014["year"]  = pd.DatetimeIndex(df_2014["birth_date"]).year
df_2010["year"]  = pd.DatetimeIndex(df_2010["birth_date"]).year
df_2006["year"]  = pd.DatetimeIndex(df_2006["birth_date"]).year
df_2002["year"]  = pd.DatetimeIndex(df_2002["birth_date"]).year
df_1998["year"]  = pd.DatetimeIndex(df_1998["birth_date"]).year
df_1994["year"]  = pd.DatetimeIndex(df_1994["birth_date"]).year
df_1990["year"]  = pd.DatetimeIndex(df_1990["birth_date"]).year
df_1986["year"]  = pd.DatetimeIndex(df_1986["birth_date"]).year
df_1982["year"]  = pd.DatetimeIndex(df_1982["birth_date"]).year
```

```python
df_2018["age"]  = 2018 - df_2018["year"]
df_2014["age"]  = 2014 - df_2014["year"]
df_2010["age"]  = 2010 - df_2010["year"]
df_2006["age"]  = 2006 - df_2006["year"]
df_2002["age"]  = 2002 - df_2002["year"]
df_1998["age"]  = 1998 - df_1998["year"]
df_1994["age"]  = 1994 - df_1994["year"]
df_1990["age"]  = 1990 - df_1990["year"]
df_1986["age"]  = 1986 - df_1986["year"]
df_1982["age"]  = 1984 - df_1982["year"]
```

```python
df_2018["year_game"] = 2018
df_2014["year_game"] = 2014
df_2010["year_game"] = 2010
df_2006["year_game"] = 2006
df_2002["year_game"] = 2002
df_1998["year_game"] = 1998
df_1994["year_game"] = 1994
df_1990["year_game"] = 1990
df_1986["year_game"] = 1986
df_1982["year_game"] = 1982
```

```python
df_18 = df_2018[['goal_keeper','defender','midfielder','forward', 'year_game','age']].reset_index() 
df_14 = df_2014[['goal_keeper','defender','midfielder','forward', 'year_game','age']].reset_index() 
df_10 = df_2010[['goal_keeper','defender','midfielder','forward', 'year_game','age']].reset_index() 
df_06 = df_2006[['goal_keeper','defender','midfielder','forward', 'year_game','age']].reset_index() 
df_02 = df_2002[['goal_keeper','defender','midfielder','forward', 'year_game','age']].reset_index() 
df_98 = df_1998[['goal_keeper','defender','midfielder','forward', 'year_game','age']].reset_index() 
df_94 = df_1994[['goal_keeper','defender','midfielder','forward', 'year_game','age']].reset_index() 
df_90 = df_1990[['goal_keeper','defender','midfielder','forward', 'year_game','age']].reset_index() 
df_86 = df_1986[['goal_keeper','defender','midfielder','forward', 'year_game','age']].reset_index() 
df_82 = df_1982[['goal_keeper','defender','midfielder','forward', 'year_game','age']].reset_index() 
```

```python
teste = pd.concat([ df_82,df_86,df_90,df_94,df_98 ,df_02, df_06 ,df_10, df_14,df_18])
```

```python
teste = teste.drop(axis=1, columns = 'index')
```
